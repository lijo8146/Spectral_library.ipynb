```python
def read_spectrum(file_path):
    try:
        with open(file_path, 'r') as file:
            # Read the entire file content
            file_content = file.read()

        # Print the content of the file
        print(file_content)

    except FileNotFoundError:
        print("Error: File not found.")
    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    file_path = r"C:\Users\lilly\Documents\ECOSTRESS spectral library\vegetation.grass.avena.fatua.tir.vh352.ucsb.nicolet.spectrum.txt"
    read_spectrum(file_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_additional_info(file_path):
    try:
        # Load data from the file using numpy
        data = np.loadtxt(file_path, delimiter='\t', skiprows=21)

        # Debug: Print the loaded data and its shape
        print("Loaded Data:")
        print(data)
        print("Data Shape:", data.shape)

        # Check if the data has at least two columns (X and Y)
        if data.shape[1] < 2:
            raise ValueError("Data must have at least two columns (X and Y).")

        # Extract additional information (X) and Y values
        additional_info = data[:, 0]  # X is in the first column
        y_values = data[:, 1]         # Y is in the second column

        # Create a plot
        plt.figure(figsize=(10, 6))
        plt.plot(additional_info, y_values, label='Y', color='green')

        # Add labels and title
        plt.xlabel('Wavelength (micrometers)')
        plt.ylabel('Reflectance (%)')
        plt.title('Plot of Spectral Signature')
        plt.legend()

        # Show the plot
        plt.grid()
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    file_path = r"C:\Users\lilly\Documents\ECOSTRESS spectral library\vegetation.grass.avena.fatua.tir.vh352.ucsb.nicolet.spectrum.txt"
    plot_additional_info(file_path)

```

```python
# third version

import numpy as np
import matplotlib.pyplot as plt
import os

def plot_additional_info(file_path, output_file_path):
    try:
        # Load data from the file using numpy
        data = np.loadtxt(file_path, delimiter='\t', skiprows=21)

        # Debug: Print the loaded data and its shape
        print("Loaded Data:")
        print(data)
        print("Data Shape:", data.shape)

        # Check if the data has at least two columns (X and Y)
        if data.shape[1] < 2:
            raise ValueError("Data must have at least two columns (X and Y).")

        # Extract additional information (X) and Y values
        additional_info = data[:, 0]  # X is in the first column
        y_values = data[:, 1]         # Y is in the second column

        # Display the file contents
        print("\nFile Contents:")
        print(np.column_stack((additional_info, y_values)))

        # Extract the plot title from the output_file_path
        plot_title = os.path.basename(output_file_path)
        plot_title = os.path.splitext(plot_title)[0]

        # Create a plot
        plt.figure(figsize=(10, 6))
        plt.plot(additional_info, y_values, label='Y', color='green')

        # Add labels and title
        plt.xlabel('Wavelength (micrometers)')
        plt.ylabel('Reflectance (%)')
        plt.title('Plot of ' + plot_title)
        plt.legend()

        # Show the plot
        plt.grid()

        # Save the plot to the specified output file path
        plt.savefig(output_file_path)

        # Close the plot to release resources
        plt.close()

    except Exception as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    input_file_path = r"C:\Users\lilly\Documents\ECOSTRESS spectral library\vegetation.tree.salix.babylonica.tir.jpl184.jpl.nicolet.spectrum.txt"
    output_file_path = r"C:\Users\lilly\Documents\salix_babylonica_2.png"
    plot_additional_info(input_file_path, output_file_path)
```
